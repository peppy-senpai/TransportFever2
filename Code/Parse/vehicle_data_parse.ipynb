# parse vehicle data

```python
import re
import os
import ast
from slpp import SLPP
import pandas as pd
from tqdm import tqdm
```

```python
def parse_mdl(filename):
    with open(filename, "r") as f:
        content = f.read()

    # Strip the function wrapper, keep only the table
    table_str = re.search(r'return\s*(\{.*\})', content, re.DOTALL).group(1)
    lua = SLPP()
    data = lua.decode(table_str)  
    return data  
```

## parse plane
### aeroplane
print(data['metadata']['availability'])     
print(data['metadata']['maintenance'])      

print(data['metadata']['airVehicle'])

print(data['metadata']['transportVehicle']['carrier'])

```python

foldername = r"C:\Users\prade\OneDrive\Documents\Code\TransportFever2\Data\vehicle\plane"
records = {
    "name":[],
    "yearFrom":[], "yearTo":[],
    "lifespan":[],
    "power": [],
    "topSpeed": [],
    "weight": [],
    "carrier":[],
    "cargo_type":[],
    "loadSpeed": [],

}

for dirpath, dirnames, filenames in os.walk(foldername):
    for filename in tqdm(filenames):
        mdl_file = os.path.join(dirpath, filename)
        data = parse_mdl(mdl_file)

        # filename
        records['name'].append(filename.split('.')[0])

        # availability
        records["yearFrom"].append(data['metadata']['availability'].get('yearFrom', None))
        records["yearTo"].append(data['metadata']['availability'].get('yearTo', None))

        # lifespan
        records["lifespan"].append(data['metadata']['maintenance'].get('lifespan', None))

        # thrust to power
        idlethrust = data['metadata']['airVehicle']['idleThrust']
        maxthrust = data['metadata']['airVehicle']['maxThrust']
        records["power"].append(0.4*idlethrust + 0.6*maxthrust)

        records['topSpeed'].append(data['metadata']['airVehicle']['topSpeed'])

        # weight
        baseWeight = data['metadata']['airVehicle']['weight']
        maxWeight = data['metadata']['airVehicle']['maxTakeOffWeight']
        records["weight"].append(0.4*baseWeight + 0.6*maxWeight)

        #carrier type
        records["carrier"].append(data['metadata']['transportVehicle']['carrier'])

        # cargo_type and capacity
        cargo_type = {}
        for cargo in data['metadata']['transportVehicle']['compartmentsList'][0]['loadConfigs']:
            cargo_type[cargo['cargoEntries'][0]['type']] = cargo['cargoEntries'][0]['capacity']
        records["cargo_type"].append(cargo_type)

        # loadSpeed
        records["loadSpeed"].append(data['metadata']['transportVehicle']['loadSpeed'])

aero_df = pd.DataFrame(records)    
aero_df['vehicle_type'] = 'plane'    
```
