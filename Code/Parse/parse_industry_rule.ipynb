# parse industry rules from .con file

```python
import re
import os
import ast
from slpp import SLPP
import pandas as pd
```

```python
# read and parse
def parse(content):
    slpp = SLPP()
    for line in content:
        if 'local stockListConfig' in line:
            found = 4
            string = [line]
            rule = {}
            while found > 0:
                line = next(content)
                string.append(line)
                found -= 1
            string = "".join(string)
            string = string.replace(r'local stockListConfig =', '')
            data = slpp.decode(string)
            return data
```

```python
# iterate through .con files
con_folder = r"C:\Users\prade\OneDrive\Documents\Code\TransportFever2\Data\game_model_file\construction\industry"
f = []
rule = []
for root, dirs, files in os.walk(con_folder):
    for file in files:
        if '.con' in file:
            filename = os.path.join(root, file)
            with open(filename, "r") as content: 
                f.append(file)   
                r = parse(content)
                rule.append(r)
            
```
